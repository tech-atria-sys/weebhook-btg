```python
import json
import requests
import uuid
import base64
import os
from dotenv import load_dotenv
from datetime import datetime, timedelta
```

```python
def save_json(resp, nome):
    with open('{}.json'.format(nome), 'w', encoding='utf-8') as f:
        json.dump(resp, f, ensure_ascii=False, indent=4)
        
def open_json(string):
    f = open(string,encoding='UTF-8')
    return json.load(f)
```

```python
try:
    load_dotenv()
except:
    load_dotenv(encoding='utf-16')

# 1. Pega o Token
def get_token_btg():
    url = "https://api.btgpactual.com/iaas-auth/api/v1/authorization/oauth2/accesstoken"
    headers = {
        'x-id-partner-request': str(uuid.uuid4()),
        'Content-Type': 'application/x-www-form-urlencoded',
        'Accept': 'application/json'
    }
    payload = {'grant_type': 'client_credentials'}
    auth = (os.getenv("btg_client_id"), os.getenv("btg_client_secret"))
    
    r = requests.post(url, data=payload, headers=headers, auth=auth)
    if r.status_code == 200:
        return r.headers.get('access_token')
    return None

def get_positions_legado(token):
    url = 'https://api.btgpactual.com/iaas-api-position/api/v1/position/partner'
    
    headers = {
        'x-id-partner-request': str(uuid.uuid4()),
        'access_token': token,
        'Accept': 'application/json',
    }
    
    print("Tentando buscar SEM especificar data (Jeito Legado)...")
    
    r = requests.get(url, headers=headers) 
    
    print(f"Status: {r.status_code}")
    print("Resposta:", r.text)

# Executa
t = get_token_btg()
if t:
    get_positions_legado(t)
```

```python
def get_positions(token):
    url='https://api.btgpactual.com/iaas-api-position/api/v1/position/partner'
    headers={
        'x-id-partner-request': str(uuid.uuid4()),
        'access_token': token,
        'Accept' : 'application/json',
    }
    r=requests.get(url, headers = headers)
    return r.json()
```

```python
def get_positions_download(token):
    url='https://api.btgpactual.com/iaas-api-position/api/v1/position/partner'
    headers={
        'x-id-partner-request': str(uuid.uuid4()),
        'access_token': token,
        'Accept' : 'application/json',
    }
    r=requests.get(url, headers = headers)
    return r.json()['response']['url']
```
